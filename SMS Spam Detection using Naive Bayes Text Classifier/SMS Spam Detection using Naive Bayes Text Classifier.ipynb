### SMS Spam Detection using Naive Bayes Text Classifier

### Problem statement

Build a text classification model using the Naive Bayes algorithm to automatically detect whether a given SMS message is spam or not. The solution uses natural language processing techniques and probabilistic learning to classify messages efficiently, aiming for high accuracy, precision, and recall.

#### Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

#### Load Dataset

```python
data = pd.read_csv('SMSSpamCollection', sep = '\t', header = None, names = ['label', 'message'])
data['label_num'] = data['label'].map({'ham' : 0, 'spam' : 1})
```

#### Split the dataset

```python
X_train, X_test, y_train, y_test = train_test_split(data['message'], data['label_num'], test_size = 0.2, random_state = 42)
```

#### Feature Extraction

```python
tfidf = TfidfVectorizer(stop_words = 'english')
X_train_vec = tfidf.fit_transform(X_train)
X_test_vec = tfidf.transform(X_test)
```

#### Train the model

```python
model = MultinomialNB(alpha = 1.0)
model.fit(X_train_vec, y_train)
```

#### Evaluate the model

```python
y_pred = model.predict(X_test_vec)
print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```
